```python
# Stock volume for every ticker

import numpy as np

table = pd.pivot_table(df, index = ['<ticker>'], values =['<vol>'], aggfunc=np.sum)

table
```

```python

BOY_Close = []
EOY_Close = []

for i in range(len(df['<ticker>'])):
    if df['<ticker>'][i] != df['<ticker>'][i + 1]:
        BOY_Close.append(df['<close>'][i + 1])
        EOY_Close.append(df['<close>'][i])
```

```python
BOY_Close.insert(0,40.94)
EOY_Close.insert(-1,0.79)
```

```python
len(BOY_Close)
```

```python
len(EOY_Close)
```

```python
BOY_EOY = list(zip(BOY_Close,EOY_Close))
```

```python
BOY_EOY
```

```python
Year_Change = []
Per_Change = []

for i,j in BOY_EOY:
    Year_Change.append(round(j-i,2))
    Per_Change.append(round((j-i)/i,4))
```

```python
Year_Change
```

```python
Per_Change
```

```python
Year_Change_df = pd.DataFrame({'Yearly_Change' : Year_Change})
Per_Change_df = pd.DataFrame({'Percent_Change' : Per_Change})
```

```python
table2 = table.reset_index()
table2
```

```python
new_df1 = table2.join(Year_Change_df)
new_df2 = new_df1.join(Per_Change_df)
```

```python
new_df3 = new_df2.rename(columns={'<ticker>':'Ticker', '<vol>':'Total_Stock_Volume'})
```

```python
new_df4 = new_df3[['Ticker', 'Yearly_Change','Percent_Change','Total_Stock_Volume']]
```

```python
def color_negative_red(value):
  """
  Colors elements in a dateframe
  green if positive and red if
  negative. Does not color NaN
  values.
  """

  if value < 0:
    color = 'red'
  elif value > 0:
    color = 'green'
  else:
    color = 'black'

  return 'color: %s' % color
```

```python
new_df5 = new_df4.style.applymap(color_negative_red, subset=['Yearly_Change','Percent_Change']).format({'Percent_Change':"{:.2%}"}).format({'Total_Stock_Volume':"{:,.0f}"})
new_df5
```
